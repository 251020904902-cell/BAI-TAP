```
gpa = float(input("Nhập điểm GPA (0 - 10): "))

if gpa < 0 or gpa > 10:
    print("Điểm không hợp lệ!")
elif gpa >= 8.5:
    print("Học lực: Xuất sắc")
elif gpa >= 7.0:
    print("Học lực: Giỏi")
elif gpa >= 5.5:
    print("Học lực: Khá")
elif gpa >= 4.0:
    print("Học lực: Trung bình")
else:
    print("Học lực: Yếu")
```

```
n = int(input("Nhập số nguyên dương n: "))
tong = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        tong += i

print("Tổng các số chẵn từ 1 đến", n, "là:", tong)
```

```
mat_khau_dung = "python123"
dem = 0

while dem < 3:
    mat_khau = input("Nhập mật khẩu: ")

    if mat_khau == mat_khau_dung:
        print("Đăng nhập thành công")
        break
    else:
        dem += 1
        print("Mật khẩu sai!")

if dem == 3:
    print("Tài khoản bị khóa")
```
