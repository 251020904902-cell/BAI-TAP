```
n = int(input("Nhập vào số nguyên dương n: "))
total = 0
for i in range(2, n+1, 2):
    total += i
print("Tổng các số chẵn nhỏ hơn hoặc bằng", n, "là:", total)
```

```

s = input("Nhập chuỗi: ")

words = s.split()

num_words = len(words)

if words:
    longest_word = max(words, key=len)
else:
    longest_word = ""

print("Số từ trong chuỗi:", num_words)
print("Từ dài nhất:", longest_word)
```

```

nums = list(map(int, input("Nhập danh sách số nguyên: ").split()))

unique_nums = []
for num in nums:
    if num not in unique_nums:
        unique_nums.append(num)

unique_nums.sort()

print("Sorted list:", unique_nums)

if len(unique_nums) >= 2:
    print("Second largest:", unique_nums[-2])
else:
    print("Không có số lớn thứ hai")
```

```
scores = list(map(float, input().split()))
valid_scores = [s for s in scores if 0 <= s <= 10]
average = sum(valid_scores) / len(valid_scores) if valid_scores else 0
passed = sum(1 for s in valid_scores if s >= 5)
print(f"Average: {average}")
print(f"Passed: {passed}")
```
